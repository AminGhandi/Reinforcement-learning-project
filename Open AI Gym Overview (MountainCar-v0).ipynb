```python
import gym
```

```python
import matplotlib.pyplot as plt
import time
```

```python
env_name ='MountainCar-v0'
env = gym.make(env_name)
```

```python
env.action_space
```

```python
def chose_action(observation):
    
    #observation
    position, velocity = observation 
    
    if -0.1 < position <0.4 :
        action = 2
        
    elif position < -0.2 and velocity < 0.0 :
        action = 0
        
    else :
        action = 1
    return action
```

```python
observation = env.reset()

for steps in range(200):
    env.render('human')
    observation,reward,done,info = env.step(chose_action(observation))
    
    print('position: ',observation[0])
    print('velocity: ',observation[1])
    print('reward: ',reward)
    print('done: ',done)
    print('info: ',info)
    
    if done:
        break
        
    time.sleep(0.001)

    
env.close()
```
